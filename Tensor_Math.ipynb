```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1, 2, 3, 4])
tensor_b = torch.tensor([5, 6, 7, 8])
```

```python
#adition
tensor_a + tensor_b
```

```python
#adition longhand
torch.add(tensor_a, tensor_b)
```

```python
#subtract
tensor_b - tensor_a
```

```python
#sub function
torch.sub(tensor_b, tensor_a)
```

```python
#multiple
tensor_a * tensor_b
```

```python
#mul longhand
torch.multiply(tensor_a, tensor_b)
```

```python
#divition
tensor_b / tensor_a
```

```python
#div longhand
torch.div(tensor_b, tensor_a)
```

```python
#remainder modulus
tensor_b % tensor_a
```

```python
#rem longhand
torch.remainder(tensor_b, tensor_a)
```

```python
#exponents / power
torch.pow(tensor_a, tensor_b)
```

```python
#another way to write longhand
tensor_a.add(tensor_b)
```

```python
#reassignment
tensor_a + tensor_b
```

```python
#tensor_a = tensor_a + tensor_b
tensor_a.add_(tensor_b)
```
